```python
import math
import re
from collections import Counter

WORD = re.compile(r"\w+")

def get_cosine(v1, v2):
    intersects = set(v1.keys()) & set(v2.keys())
    num = sum([v1[x] * v2[x] for x in intersects])
    sum1 = sum([v1[x] ** 2 for x in list(v1.keys())])
    sum2 = sum([v2[x] ** 2 for x in list(v2.keys())])
    den = math.sqrt(sum1) * math.sqrt(sum2)
    if not den:
        return 0.0
    else:
        return round(float(num) / den,3)

def text_to_vector(text):
    words = WORD.findall(text)
    return Counter(words)
```

```python
all_sentences = filtered_train_en['processed_title']
all_sentences_vectors = []
for k in tqdm(range(len(all_sentences))):
    all_sentences_vectors.append(text_to_vector(all_sentences.iloc[k]))
```

```python
all_sentences_vectors
```

```python
all_sample_vectors = []
for n in tqdm(range(len(translation_result))):
    all_sample_vectors.append(text_to_vector(translation_result[n]))
```

```python
len(all_sample_vectors)
```

Parallel mini-batch processing of translation by Google translation API

```python
doc = []
for j in tqdm(range(120000,122000)): #set your mini-batch range here
    cosine_similarity = []
    vector1 = all_sample_vectors[j]
    for i in range(len(all_sentences)):
        vector2 = all_sentences_vectors[i]
        cosine_similarity.append(get_cosine(vector1, vector2))
    doc.append([filtered_train_tcn['product_title_tokenized'].iloc[j], translation_result[j], all_sentences.iloc[cosine_similarity.index(max(cosine_similarity))], max(cosine_similarity)])

print('Cosine Similarity calculation completed')
```

```python
df = pd.DataFrame(doc, columns = ['Original','Translated','Matched','Cosine_sim'])
df.to_csv('train_matched64.csv', index=False)
```
