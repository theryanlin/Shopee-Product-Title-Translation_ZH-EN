```python
import pandas as pd
import numpy as np
import re

train = pd.read_csv('./datasets/train_en.csv')
dev = pd.read_csv('./datasets/dev_en.csv')
```

1. punctuation removal 
2. numbers removal 
3. emoji/special characters removal
4. transfer to lowercase
5. tokenization
6. stopwords removal (ignore)

```python
def processor(text):
    text = re.sub(r'[^\w]', ' ', text.lower()) # remove non english words, special characters
    text = re.sub(r'\b\w*[\d]+\w*\b', ' ', text) # remove non digits
    text = ' '.join(text.split()) # remove extra spaces
    return text
```

```python
train['processed_title'] = train['product_title'].apply(processor)
dev['translation_output'] = dev['translation_output'].apply(processor)
```
